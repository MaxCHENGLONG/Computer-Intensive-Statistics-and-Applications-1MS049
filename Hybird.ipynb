# 生成随机数的过程

```python
# 如果想要是random_numbers1独一无二， randm_numbers2是有重复的值，且大约从random_numbers1中有放回采样4000 则 Flag = 1
# 如果两个都是独一无二的，且大约从random_numbers1中无放回采样4000 则 Flag = 2
import numpy as np
import random  
seed = 3407
random.seed(seed)
np.random.seed(seed)
Flag = 3  # 1 代表着 A 列是独一无二 B列是有重复的值， 2 代表着两列都是独一无二的 3代表含有MCV的情况
N = 10000 # 数据总数
#radom_nk = np.random.randint(3000,6000)
#NK = radom_nk # 两个表之间相同的个数
NK = 4637
fanwei = 20000+1
N_100 = np.random.randint(1, 30) # 100个数中有多少个数是不重复的

if Flag == 1:
    import random 
    import numpy as np
    random_numbers1 = random.sample(range(1, fanwei), N)
    random_numbers1.sort()
    pre_random_numbers2 = random.choices(random_numbers1, k=NK) # 有放回采样
    random_numbers2 = pre_random_numbers2 + random.sample(range(1, fanwei), N - NK) #这个情况是加入了一些随机性，即给B列中加入一个重复的值
    random_numbers2.sort()
    print(f"random_numbers1 中的总数: {len(random_numbers1)}")
    print(f"random_numbers2 中的总数: {len(random_numbers2)}")
elif Flag == 2:
    import random
    import numpy as np
    # 已经生成的数据
    random_numbers1 = random.sample(range(1, fanwei), N)
    random_numbers1.sort()
    pre_random_numbers2 = random.sample(random_numbers1, k=NK)
    # 将 random_numbers1 和 pre_random_numbers2 转为集合以便快速操作
    set_random_numbers1 = set(random_numbers1)
    set_pre_random_numbers2 = set(pre_random_numbers2)

    # 剩余的候选数范围
    remaining_candidates = set(range(1, fanwei)) - set_random_numbers1 - set_pre_random_numbers2

    # 从剩余候选数中选择所需的数
    remaining_random_numbers2 = random.sample(remaining_candidates, k=(N - len(pre_random_numbers2)))

    # 将 pre_random_numbers2 与剩余数结合，形成完整的 random_numbers2
    random_numbers2 = list(pre_random_numbers2) + remaining_random_numbers2

    # 如果需要排序（可选）
    random_numbers2.sort()
    print(f"random_numbers1 中的总数: {len(random_numbers1)}")
    print(f"random_numbers2 中的总数: {len(random_numbers2)}")
elif Flag == 3:
    import random 
    import numpy as np
    set1 = random.sample(range(1, fanwei), N) # 无重复值的随机数 1～200001 之间 抽取10000个数
    random_numbers1 = set1
    set2 = set()
    while len(set2) < N:
        num = random.randint(1, fanwei)
        set2.add(num)
    random_numbers2 = list(set2)
    repeated_numbers = []
    non_overlapping_numbers = []
    for _ in range(N_100):
        num = random.choice([x for x in range(1, fanwei) if x not in set1])
        count = random.randint(10, 100)  # 每个数的重复次数
        non_overlapping_numbers.extend([num] * count)

    repeated_numbers.extend(non_overlapping_numbers)
    # 生成100-N_100其他重复数
    for _ in range(100-N_100):
        num = random.randint(1, fanwei)
        count = random.randint(10, 100)
        repeated_numbers.extend([num] * count)

    set2 = list(set2)
    set2.extend(repeated_numbers)
    random.shuffle(set2)

    # 调整以满足4000个左右的重复项
    set1_set2_intersection = len(set(set1) & set(set2))
    while set1_set2_intersection < 4000:
        num = random.choice(set1)
        set2.append(num)
        set1_set2_intersection += 1
    # Remove elements from set2 if they appear only once and not in set1 until lengths match
    # count_dict = {num: set2.count(num) for num in set2}
    # set2 = [num for num in set2 if num in set1 or count_dict[num] > 1]

    # # If set2 is still longer than set1, trim it down
    # while len(set2) > len(set1):
    #     set2.pop()
    random_numbers1 = set1
    random_numbers2 = set2
    print(f"第一个集合大小: {len(set1)}")
    print(f"第二个集合大小: {len(set2)}")
    print(f"相同数字个数: {set1_set2_intersection}")


```

# 构建混合直方图

```python
# M是桶的个数
M = 254 # number of bins

N1 = len(random_numbers1)
N2 = len(unique_numbers)
if N1 % M == 0:
    print("N1 可以被 M 整除")
else:
    print("N1 不能被 M 整除")


if N2 % M == 0:
    print("N2 可以被 M 整除")
else:
    print("N2 不能被 M 整除")
```

### 等高直方图的创建过程

```python
N1 = len(random_numbers1)  # 总数据长度
N2 = len(unique_numbers)
# 检查是否有足够的数据进行分 bin
if M > N1:
    raise ValueError("M 不能大于数据的总长度")

if M > N2:
    raise ValueError("M 不能大于数据的总长度")
bin_size1 = N1 // M  # 每个 bin 的大小
remainder1 = N1 % M  # 剩余数据
bin_size2 = N2 // M  # 每个 bin 的大小
remainder2 = N2 % M  # 剩余数据
# 判断是否有剩余数据
if remainder1 == 0:
    # 无剩余数据，正常分配
    bins1 = {f"bin_{i+1}": random_numbers1[i*bin_size1:(i+1)*bin_size1] for i in range(M)}
    #bins2 = {f"bin_{i+1}": random_numbers2[i*bin_size:(i+1)*bin_size] for i in range(M)}
else:
    # 有剩余数据，最后一个 bin 添加剩余部分
    bins1 = {f"bin_{i+1}": random_numbers1[i*bin_size1:(i+1)*bin_size1] for i in range(M-1)}
    bins1[f"bin_{M}"] = random_numbers1[(M-1)*bin_size1:]  # 包含剩余数据

    #bins2 = {f"bin_{i+1}": random_numbers2[i*bin_size:(i+1)*bin_size] for i in range(M-1)}
    #bins2[f"bin_{M}"] = random_numbers2[(M-1)*bin_size:]  # 包含剩余数据


# 判断是否有剩余数据
if remainder2 == 0:
    # 无剩余数据，正常分配
    #bins1 = {f"bin_{i+1}": random_numbers1[i*bin_size:(i+1)*bin_size] for i in range(M)}
    bins2 = {f"bin_{i+1}": unique_numbers[i*bin_size2:(i+1)*bin_size2] for i in range(M)}
else:
    # 有剩余数据，最后一个 bin 添加剩余部分
    #bins1 = {f"bin_{i+1}": random_numbers1[i*bin_size:(i+1)*bin_size] for i in range(M-1)}
    #bins1[f"bin_{M}"] = random_numbers1[(M-1)*bin_size:]  # 包含剩余数据

    bins2 = {f"bin_{i+1}": unique_numbers[i*bin_size2:(i+1)*bin_size2] for i in range(M-1)}
    bins2[f"bin_{M}"] = unique_numbers[(M-1)*bin_size2:]  # 包含剩余数据

# 打印检查
print(f"Bins1: {bins1}")
print(f"Bins2: {bins2}")
print(f"bin1是否有剩余数据: {'是' if remainder1 > 0 else '否'}")
print(f"bin2是否有剩余数据: {'是' if remainder2 > 0 else '否'}")
```

## 计算 估计的行数

```python
ans_fre = 0
for j in range(len(duplicates_with_counts)):
    for i in range(1, M + 1):
        # 获取 bin 的左右端点
        bins1_right = max(bins1[f"bin_{i}"])  # A表中第i个bin的右端点值
        bins1_left = min(bins1[f"bin_{i}"])   # A表中第i个bin的左端点值
        print("bins1_left:", bins1_left)
        print("bins1_right:", bins1_right)
        
        # 检查当前数字是否在 bin 范围内
        current_number = duplicates_with_counts['number'].iloc[j]
        if current_number < bins1_left or current_number > bins1_right:
            continue
        else:
            print(f"第{i}个bin中包含了 {current_number}")
            # 更新 ans_fre，防止索引错误并修复表达式
            #ans_fre += min(len(bins1[f"bin_{i}"]) / (bins1_right - bins1_left), 1) * duplicates_with_counts['fre'].iloc[j] * len(random_numbers2)
            ans_fre += min(len(bins1[f"bin_{i}"]) / (bins1_right - bins1_left), 1) * duplicates_with_counts['count'].iloc[j]

ans_est = 0
for i in range(1,M+1):
    for j in range(1,M+1):
        bins1_right = max(bins1[f"bin_{i}"]) # A表中第i个bin的右端点值
        
        bins1_left = min(bins1[f"bin_{i}"]) # A表中第i个bin的左端点值
        
        bins2_right = max(bins2[f"bin_{j}"]) # B表中第j个bin的右端点值
        
        bins2_left = min(bins2[f"bin_{j}"]) # B表中第j个bin的左端点值

        if bins1_right < bins2_left or bins1_left > bins2_right:
            continue
        print("bins1_left:", bins1_left)
        print("bins1_right:", bins1_right)
        print("bins2_left:", bins2_left)
        print("bins2_right:", bins2_right)
        right_over = min(bins1_right, bins2_right)
        
        left_over = max(bins1_left, bins2_left)
        print("left_over:", left_over)
        print("right_over:", right_over)
        bins1_pa = (right_over - left_over)/(bins1_right - bins1_left) * len(bins1[f"bin_{i}"])
        bins2_pa = (right_over - left_over)/(bins2_right - bins2_left) * len(bins2[f"bin_{j}"])
        if right_over == left_over:
            ans_est = ans_est + bins2[f'bin_{j}'].count(bins2[f'bin_{j}'][-1])
        elif bins1_right == right_over:
            print("bins1_right == right_over")
            temp = (bins1_pa * bins2_pa)/(right_over - left_over)
            print(f"第{i}个bin和第{j}个bin的交集的估计值为: {temp}")
            ans_est = ans_est + temp 
        elif bins2_right == right_over:
            print("bins2_right == right_over")
            temp =(bins1_pa * bins2_pa)/(right_over - left_over)
            print(f"第{i}个bin和第{j}个bin的交集的估计值为: {temp}")
            ans_est = ans_est + temp 
        else:
            temp = (bins1_pa * bins2_pa)/(right_over - left_over)
            print(f"第{i}个bin和第{j}个bin的交集的估计值为: {temp}")
            ans_est = ans_est + temp 
        print("\n")
hybird_result = ans_est + ans_fre
print("MCV:",ans_fre)
print("Non-MCV:",ans_est)
print("hybird_result:", hybird_result)
#print(ans_fre)
```
