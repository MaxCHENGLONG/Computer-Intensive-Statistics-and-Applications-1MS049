```python
import random

# 设置范围和初始随机数数量
total_numbers = 10000
range_start = 1
range_end = 200000

# 生成不重复的随机数
random_numbers1 = random.sample(range(range_start, range_end + 1), total_numbers)

# 从 random_numbers1 中随机选择 2000 个数字
random_numbers2 = random.sample(random_numbers1, 2000)

# 剩余的 8000 个数字
remaining_numbers = list(set(random_numbers1) - set(random_numbers2))

# 生成新的 8000 个随机数，确保不与 random_numbers1 重复
new_random_numbers = random.sample(
    list(set(range(range_start, range_end + 1)) - set(random_numbers1)),
    8000
)
random_numbers2 += new_random_numbers
# 验证
print(f"random_numbers2 数量: {len(random_numbers2)}")
print(f"new_random_numbers 数量: {len(new_random_numbers)}")
print("部分 random_numbers2: ", random_numbers2[:10])  # 打印部分 random_numbers2
print("部分 new_random_numbers: ", new_random_numbers[:10])  # 打印部分 new_random_numbers
```

```python
import numpy as np

# 确保数据已排序
random_numbers1.sort()
random_numbers2.sort()

# 将数据分为 256 份（整数边界）
bins1 = np.linspace(min(random_numbers1), max(random_numbers1), 257, dtype=int)
bins2 = np.linspace(min(random_numbers2), max(random_numbers2), 257, dtype=int)

# 统计每个 bin 的 count
def count_in_bins(data, bins):
    bin_counts = []
    for i in range(len(bins) - 1):
        bin_data = [x for x in data if bins[i] <= x < bins[i + 1]]
        if bin_data:  # 确保 bin 不为空
            left, right = min(bin_data), max(bin_data)
            count = len(bin_data)
            bin_counts.append((left, right, count))
    return bin_counts

# 计算 bin 的分布
bin_data1 = count_in_bins(random_numbers1, bins1)
bin_data2 = count_in_bins(random_numbers2, bins2)

# 计算重复值的估计
total_count1 = len(random_numbers1)
total_count2 = len(random_numbers2)

# 存储重复值的估计
overlap_estimates = []
for i, (left1, right1, count1) in enumerate(bin_data1):
    for j, (left2, right2, count2) in enumerate(bin_data2):
        # 计算两个 bin 的重叠区间
        overlap_left = max(left1, left2)
        overlap_right = min(right1, right2)

        if overlap_left < overlap_right:  # 如果有重叠
            overlap_length1 = (overlap_right - overlap_left) / (right1 - left1) 
            overlap_length2 = (overlap_right - overlap_left) / (right2 - left2) 

            # 计算重复值估计
            estimate = (
                (overlap_length1 * count1 + overlap_length2 * count2) / (overlap_right - overlap_left)
            )
            overlap_estimates.append((i, j, estimate))  # 保存 bin 索引和估计值

# 打印结果
ans_estimates = 0
for i, j, estimate in overlap_estimates:
    ans_estimates += estimate
    print(f"Bin1[{i}] overlaps Bin2[{j}] with estimate: {estimate}")
```
